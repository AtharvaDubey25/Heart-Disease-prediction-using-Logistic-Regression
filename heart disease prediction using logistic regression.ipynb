```python
from sklearn.impute import KNNImputer
```

```python
smokers=df[(df['currentSmoker']==1)].copy()
nonsmokers=df[(df['currentSmoker']==0)].copy()
```

```python
imputer=KNNImputer(n_neighbors=3)
```

```python
smokers_imputed=imputer.fit_transform(smokers.drop(columns=['currentSmoker']))
smokers=pd.DataFrame(smokers_imputed,columns=smokers.drop(columns=['currentSmoker']).columns)
```

```python
nonsmokers['cigsPerDay']=nonsmokers['cigsPerDay'].fillna(0.0)
nonsmokers=nonsmokers.drop(columns=['currentSmoker'])
```

```python
df_new=pd.concat([smokers,nonsmokers],axis=0).sort_index()
```

```python
df_new_=pd.DataFrame(df_,columns=df_new.columns)
```

```python
df_new_.drop(columns='male',inplace=True)
```

```python
X=df_new_.iloc[:,:-1]
y=df_new_.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sclr=StandardScaler()
```

```python
new_df=sclr.fit_transform(X[['age','cigsPerDay','totChol','sysBP','diaBP','BMI','heartRate','glucose']])
```

```python
new_df.shape
```

```python
new_df_=pd.DataFrame(new_df,columns=['age','cigsPerDay','totChol','sysBP','diaBP','BMI','heartRate','glucose'])
```

```python
new_df_.head()
```

```python
X.drop(columns=['age','cigsPerDay','totChol','sysBP','diaBP','BMI','heartRate','glucose'],inplace=True)
```

```python
X.head()
```

```python
X=pd.concat([new_df_,X,df[['male']]],axis=1)
```

```python
X.head()
```

```python
X.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2,test_size=0.3)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix,recall_score
```

```python
f1_score(y_test,y_pred)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```
